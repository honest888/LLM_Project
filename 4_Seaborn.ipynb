# SEABORN

Seaborn is a statistical plotting lib, built on matplotlib

## Distribution plot

```python
import seaborn as sns
%matplotlib inline
```

```python
#builtin dataset in seaborn
tips = sns.load_dataset('tips')
tips.head()
```

 **DIST PLOT**
 this allows us to show distribution of a univariate set of operations(just one variable)

```python
sns.displot(tips['total_bill'],bins=40)
```

```python
sns.displot(tips['total_bill'],kde=True)
```

## Joint plot
combines 2 diff dsitribution plot(2variable)

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
#hex/scatter sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex/reg/kde(density)') 
```

```python
#quickly visualizing our data

sns.pairplot(tips,hue='sex',palette='coolwarm')

#sns.pairplot(tips)
```

## Rugplot
Draws dash marks for every point for any one variable

```python
sns.rugplot(tips['total_bill'])
```

# Categorical plot

```python
tips.head()
```

```python
import numpy as np
```

```python
#barplot lets u aggregate the categorical data w a function(like mean)
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#by default estimaor is mean
```

## Count plot
same as bar ploy, but estimator is explicitly counting the no. of occurences, because of that we set only x

```python
sns.countplot(x='sex',data=tips)
```

## Box Plot

Shows th distribution of quantitative data in a way that faciliates the comparison in btween variable

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#the dots are outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

## VIOLIN PLOT

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Strip plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
#jitter-> density
```

# MATRIX PLOTS

Heat maps!!

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

For a heat map to work properly, **the dataset** must be in **matrix form**  So that the index name and col name match up so that the cell value actually indicates something that is relevant to both of these name.


```python
tc=tips.corr()
tc
#since w needed both variables on the cols and rows
#this could have been done using a pivot table or correlation data
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
#pivot form 
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp)
```

```python
sns.clustermap(fp)
#clusters info to show common row n col
```
