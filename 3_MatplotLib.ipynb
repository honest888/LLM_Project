# MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.linspace(0,5,11)
#0-5 and 11 points
y=x**2
```

```python
x
```

```python
y
```

**we can us matplotlib in two ways- functions and object oriented**

## Functional 

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'red')
#or plt.plot(x,y,'-r')
```

```python
#labels
plt.plot(x,y,'-r')
plt.xlabel('X LABEL')
plt.ylabel('Y LABEL')
plt.title('TITLE')
```

```python
plt.subplot(1,2,1)
#rowss,col,plot no. u referring to
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
#rowss,col,plot no. u referring to
plt.plot(y,x,'b*-')
```

# OBJECT ORIENTED

```python
#create fig obj
fig=plt.figure()
#it takes 4 arg, by passing a list (left,bottom,width,height arg) from 0-1
axis = fig.add_axes([0.1,0.1,0.8,0.8])

#plotting aixs
axis.plot(x,y)
axis.set_xlabel('X LABEL')
axis.set_ylabel('Y LABEL')
axis.set_title('TITLE')
```

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis2=fig.add_axes([0.5,0.25,0.4,0.5])
axis1.plot(x,y,'blue')
axis2.plot(y,x,'red')
```

```python
#SUBPLOTS
fig,axes=plt.subplots(nrows=2,ncols=2)
#axes.plot(x,y)
#padding
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#iterate through graphs
for current in axes:
    current.plot(x,y)
```

```python
axes
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
#index through graphs
axes[0].plot(x,y)
axes[0].set_title('First plot!!')
axes[1].plot(y,x)

plt.tight_layout()
```

## Figure Size and DPI

```python
fig=plt.figure(figsize=(4,3),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(4,3))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#save a fig to file
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x label')
ax.plot(y,x,label='y label')

ax.legend()
#to prevent overlapping
#ax.legend(loc=0 or 10 or bla bla)
```

# setting color

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,'#ff8c00',linewidth=2,alpha=0.5,linestyle='--',marker='*',markersize=20,markerfacecolor='black') #rgb hex code
```
